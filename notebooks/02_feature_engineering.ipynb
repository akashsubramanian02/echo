```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle
import os
```

```python
DATA_PATH = "../data/clean_reviews.csv"

df = pd.read_csv(DATA_PATH)

df.head()
```

```python
X_text = df["clean_review"]
y = df["sentiment"]

print(X_text.shape, y.shape)
```

```python
vectorizer = TfidfVectorizer(
    max_features=5000,
    ngram_range=(1,2),      # improves accuracy
    stop_words="english"
)

X = vectorizer.fit_transform(X_text)

X.shape
```

```python
os.makedirs("../models", exist_ok=True)

with open("../models/tfidf_vectorizer.pkl", "wb") as f:
    pickle.dump(vectorizer, f)

print("✅ TF-IDF vectorizer saved")
```

```python
pd.DataFrame(X.toarray()).to_csv("../data/X_features.csv", index=False)
y.to_csv("../data/y_labels.csv", index=False)

print("✅ Features and labels saved")
```
